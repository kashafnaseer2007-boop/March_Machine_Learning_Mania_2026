```python
import pandas as pd
import numpy as np
```

```python
# Load test IDs
test_original = pd.read_csv('/kaggle/input/competitions/playground-series-s6e3/test.csv')
test_ids = test_original['id']

# Load your three submission files
# (You'll need to upload these to the notebook's input)
xgb_sub = pd.read_csv('/kaggle/input/datasets/kashafrana2007/my-xgb-lgb-submissions/xgb_submission.csv')
lgb_sub = pd.read_csv('/kaggle/input/datasets/kashafrana2007/my-xgb-lgb-submissions/lgb_submission.csv')
cat_sub = pd.read_csv('/kaggle/input/datasets/kashafrana2007/my-xgb-lgb-submissions/catboost_submission.csv')

# Verify IDs match
assert (xgb_sub['id'] == test_ids).all(), "XGBoost ID mismatch"
assert (lgb_sub['id'] == test_ids).all(), "LightGBM ID mismatch"
assert (cat_sub['id'] == test_ids).all(), "CatBoost ID mismatch"
```

```python
simple_avg = (xgb_sub['Churn'] + lgb_sub['Churn'] + cat_sub['Churn']) / 3

simple_sub = pd.DataFrame({'id': test_ids, 'Churn': simple_avg})
simple_sub.to_csv('ensemble_3_simple.csv', index=False)
print("Simple average ensemble saved.")
```

```python
# Replace these with your actual CV scores
xgb_cv = 0.916159658315072
lgb_cv = 0.9163886852510231
cat_cv = 0.9162142516992352 

total = xgb_cv + lgb_cv + cat_cv
w_xgb = xgb_cv / total
w_lgb = lgb_cv / total
w_cat = cat_cv / total

print(f"Weights: XGB = {w_xgb:.4f}, LGB = {w_lgb:.4f}, Cat = {w_cat:.4f}")

weighted_avg = w_xgb * xgb_sub['Churn'] + w_lgb * lgb_sub['Churn'] + w_cat * cat_sub['Churn']

weighted_sub = pd.DataFrame({'id': test_ids, 'Churn': weighted_avg})
weighted_sub.to_csv('ensemble_3_weighted.csv', index=False)
print("Weighted average ensemble saved.")
```

```python
corr = np.corrcoef(simple_avg, weighted_avg)[0,1]
print(f"Correlation between simple and weighted ensembles: {corr:.6f}")
```
